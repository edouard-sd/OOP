### Ex 1 : Nombres Complexes
Implémentation de l'addition et de la soustraction via `__add__` et `__sub__`.

```python
class Complexnumber: 
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary 

    def __add__(self, other):
        real = self.real + other.real
        imaginary = self.imaginary + other.imaginary
        return Complexnumber(real, imaginary)
    
    def __sub__(self, other):
        real = self.real - other.real
        imaginary = self.imaginary - other.imaginary 
        return Complexnumber(real, imaginary)
    
    def __str__(self):
        return f"{self.real} + {self.imaginary}i"
```

```python
# Here I created two complex numbers
c1 = Complexnumber(3, 4)
c2 = Complexnumber(1, 2)

# Here I tested addition
result_add = c1 + c2
print(f"The sum of your two complex numbers is: {result_add}")

# Here I tested subtraction
result_sub = c1 - c2
print(f"The difference of your two complex numbers is: {result_sub}")
```

**Notes :** 
- Les méthodes spéciales (dunder methods) permettent d'utiliser les opérateurs standards (`+`, `-`) sur nos objets.
- `__str__` gère l'affichage personnalisé.

### Ex 3 : Produit scalaire de vecteurs
Surcharge de `*` (`__mul__`) pour calculer le produit scalaire 3D.

```python
class Vector: 
    def __init__(self, x, y, z):
        self.x = x
        self.y = y 
        self.z = z
    
    def __mul__(self, other):
        dot_product = self.x * other.x + self.y * other.y + self.z * other.z
        return dot_product
```

```python
vector1 = Vector(1, 2, 3)
vector2 = Vector(4, 5, 6)

scalar_product = vector1.__mul__(vector2) # Here I tested scalar product
print(f"The scalar product of the two vectors is: {scalar_product}")
```

**Essentiel :** On peut retourner n'importe quel type (ici un scalaire) depuis une surcharge d'opérateur.

### Ex 4 : Opérateurs de comparaison (Cercle)
Implémentation complète : `<`, `>`, `==`, `!=`, `<=`, `>=` basée sur le rayon.

```python
class Circle: 

    def __init__(self, radius):
        self.radius = radius 

    def __lt__(self, other): # Here I defined less than
        return self.radius < other.radius 

    def __gt__(self, other): # Here I defined greater than
        return self.radius > other.radius
    
    def __eq__(self, other): # Here I defined equals
        return self.radius == other.radius
    
    def __ne__(self, other): # Here I defined not equals
        return self.radius != other.radius
    
    def __le__(self, other): # Here I defined less than or equal
        return self.radius <= other.radius 
    
    def __ge__(self, other): # Here I defined greater than or equal    
        return self.radius >= other.radius      
```

```python
Circle1 = Circle(5)
Circle2 = Circle(10)
Circle3 = Circle(2)
```

```python
print (Circle1 < Circle2) # Here I tested less than
print (Circle1 > Circle3) # Here I tested greater than
print (Circle1 == Circle3) # Here I tested equals
print (Circle1 != Circle3) # Here I tested not equals
print (Circle1 <= Circle2) # Here I tested less than or equal
print (Circle1 >= Circle3) # Here I tested greater than or equal
```

### Ex 5 : Distance Euclidienne (Point 2D)
Calcul de distance par rapport à l'origine et logique de comparaison associée.

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def Euclidean_distance(self):
       return (self.x ** 2 + self.y ** 2) ** 0.5
```

```python
point1 = Point(3, 4)
point2 = Point(6, 8)
print(f"The Euclidean distance of point 1 from the origin is: {point1.Euclidean_distance()}")
print(f"The Euclidean distance of point 2 from the origin is: {point2.Euclidean_distance()}")  

if point1.Euclidean_distance() < point2.Euclidean_distance():
    print("Point 1 is closer to the origin than Point 2")
elif point1.Euclidean_distance() > point2.Euclidean_distance():
    print("Point 2 is closer to the origin than Point 1")
else:
    print("Both points are equidistant from the origin")
```

### Ex 6 : Multiplication de matrices
Algorithme de produit matriciel avec validation des dimensions (`cols` vs `rows`).

```python
class Matrices:

    def __init__(self, data):
        self.data = data
        self.rows = len(data)
        self.cols = len(data[0]) 

    def __mul__(self, other):
        if self.cols != other.rows:
            raise ValueError("Cannot Multiply the two matrices because the number of columns in the first matrix is not equal to the number of rows in the second matrix.")
        result = [[0 for _ in range(other.cols)] for _ in range(self.rows)]
        for i in range(self.rows):
            for j in range(other.cols):
                for k in range(self.cols):
                    result[i][j] += self.data[i][k] * other.data[k][j]
        return Matrices(result)
    
```

```python
M1 = Matrices([[1, 2], [3, 4]])
M2 = Matrices([[5, 6], [7, 8]]) 
M3 = M1 * M2
print(f"The result of multiplying the two matrices is: {M3.data}")
```

## Partie 2 : Héritage

### Ex 1 : Véhicules
Utilisation de l'héritage simple pour spécialiser le comportement de `drive()`.

```python
class Vehicle: 

    def __init__(self):
        pass

    def __drive__(self):
        print("Driving the vehicle...")
```

```python
v1 = Vehicle()
v1.__drive__()
```

```python
# Here I defined Car and Bicycle subclasses

class Car(Vehicle):

    def __drive__(self):
        print("Driving the car...")

class Bicycle(Vehicle):

    def __drive__(self):
        print("Riding the bicycle...")
```

```python
c1 = Car()
c1.__drive__()

b1 = Bicycle()
b1.__drive__()
```

**Concept :** Les classes enfants héritent des méthodes du parent mais peuvent les redéfinir (polymorphisme).

### Ex 2 : Personnes et Étudiants
Utilisation de `super().__init__()` pour étendre les constructeurs des classes dérivées.

```python
class Person:

    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

    def __introduyce__(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old and I live at {self.address}.")
```

```python
person1 = Person("John", 30, "123 Main St")
person1.__introduyce__()
```

```python
class Student(Person):

    def __init__(self, name, age, address, field_of_study):
        super().__init__(name, age, address)
        self.field_of_study = field_of_study

    def __introduce__(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old and I live at {self.address}. I study {self.field_of_study}.")
```

```python
student1 = Student("John", 30, "123 Main St", "Computer Science")
student1.__introduce__()
```

```python
class Employee(Person):

    def __init__(self, name, age, address, job_title):
        super().__init__(name, age, address)
        self.job_title = job_title

    def __introduce__(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old and I live at {self.address}. I work as a {self.job_title}.")
```

```python
employee1 = Employee("Jean", 28, "Paris", "Software Engineer at Apple")
employee1.__introduce__()
```

### Ex 3 : Comptes Bancaires
Gestion d'un compte épargne héritant des fonctionnalités de base (dépôt, retrait) avec ajout d'intérêts.

```python
class BankAccount:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def __deposit__(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")

    def __withdraw__(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")

    def __transfer__(self, amount, other_account):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            other_account.balance += amount 
            print(f"Transferred {amount} to account {other_account.account_number}. New balance is {self.balance}.")

```

```python
BankAccount1 = BankAccount("123456789", 1000)
BankAccount1.__deposit__(500)
```

```python
BankAccount1.__withdraw__(200)
```

```python
class SavingsAccount(BankAccount):

    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def __add_interest__(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Added interest of {interest}. New balance is {self.balance}.")

    def __Check_balance__(self):
        print(f"Current balance is {self.balance}.")
```

```python
SavingsAccount1 = SavingsAccount("987654321", 2000, 0.05)
SavingsAccount1.__add_interest__()
```

```python
SavingsAccount1.__Check_balance__()
```

```python
BankAccount2 = BankAccount("111111111", 500)
BankAccount1.__transfer__(300, BankAccount2)
```

```python
BankAccount1.__transfer__(2000, BankAccount2)
print(f"Bank Account 1 balance: {BankAccount1.balance}")
```
